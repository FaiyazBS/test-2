```python
import pandas as pd 
from sklearn.preprocessing import StandardScaler
import os

df = pd.read_csv("/Users/faiyazbinshafiq/Playground/ML/Projects/fc26_players_position_playing_style_prediction/data/FC26_20250921.csv")
df["primary_position"] = df["player_positions"].str.split(",").str[0].str.strip()

# Detailed attribute columns actually present in this dataset
outfield_attrs = [
    "attacking_crossing","attacking_finishing","attacking_heading_accuracy",
    "attacking_short_passing","attacking_volleys",
    "skill_dribbling","skill_curve","skill_fk_accuracy","skill_long_passing","skill_ball_control",
    "movement_acceleration","movement_sprint_speed","movement_agility","movement_reactions","movement_balance",
    "power_shot_power","power_jumping","power_stamina","power_strength","power_long_shots",
    "mentality_aggression","mentality_interceptions","mentality_positioning","mentality_vision",
    "mentality_penalties","mentality_composure",
    "defending_marking_awareness","defending_standing_tackle","defending_sliding_tackle"
]
goalkeeping_attrs = [
    "goalkeeping_diving","goalkeeping_handling","goalkeeping_kicking",
    "goalkeeping_positioning","goalkeeping_reflexes","goalkeeping_speed"
]
core_attrs = ["pace","shooting","passing","dribbling","defending","physic"]

feature_cols = core_attrs + outfield_attrs + goalkeeping_attrs

# Fill blanks meaningfully, not just with the column mean
df[core_attrs] = df[core_attrs].fillna(0)
df[outfield_attrs] = df[outfield_attrs].fillna(0)
df[goalkeeping_attrs] = df[goalkeeping_attrs].fillna(0)

X = df[feature_cols]
y_position = df["primary_position"]

#Scaling for K-means
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#Saving into disk
os.makedirs("/Users/faiyazbinshafiq/Playground/ML/Projects/fc26_players_position_playing_style_prediction/data/processed", exist_ok=True)

# id/name columns kept for reference, feature_cols kept for modeling, primary_position as the label
keep_cols = ["player_id", "short_name", "primary_position"] + feature_cols
df[keep_cols].to_csv("/Users/faiyazbinshafiq/Playground/ML/Projects/fc26_players_position_playing_style_prediction/data/processed/fc26_preprocessed.csv", index=False)
```
